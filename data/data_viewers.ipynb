## Detention data quick analysis
- Load detention file
- Check duplicates by `Unique Identifier` (possible transfers)
- Year/month counts (if a date column exists)
- Sex proportions
- Age median (overall/by year)
- Notes will print if a column is missing


```python
import polars as pl
from pathlib import Path

# Paths
base_dir = Path("../data")
DETENTIONS_PATH = base_dir / "raw/ice_release_11aug2025_with_removals/2025-ICLI-00019_2024-ICFO-39357_ICE Detentions_LESA-STU_FINAL Redacted_raw.xlsx"

print("Loading detentions from", DETENTIONS_PATH)
df = pl.read_excel(DETENTIONS_PATH, engine="calamine", read_options={"header_row": 6})
print("Rows, cols:", df.shape)
print("Columns:", df.columns)
print("Head:")
print(df.head(5))

# Explicit column selection based on inspection
uid_col = "Unique Identifier" if "Unique Identifier" in df.columns else None
sex_col = "Gender" if "Gender" in df.columns else None
birth_year_col = "Birth Year" if "Birth Year" in df.columns else None
# Use Stay Book In Date Time as main date field
raw_date_col = "Stay Book In Date Time" if "Stay Book In Date Time" in df.columns else None

# Duplicate analysis (possible transfers)
if uid_col:
    dup_counts = df.group_by(uid_col).len().sort("len", descending=True)
    dup_summary = {
        "total_unique": dup_counts.shape[0],
        "multi_records": dup_counts.filter(pl.col("len") > 1).shape[0],
        "max_records_per_uid": dup_counts["len"].max(),
        "mean_records_per_uid": float(dup_counts["len"].mean()),
    }
    print("\nDuplicate summary (by Unique Identifier):", dup_summary)
    print("Top duplicate examples:")
    print(dup_counts.filter(pl.col("len") > 1).head(10))
else:
    print("\nNo Unique Identifier column found.")

# Parse date if present
if raw_date_col:
    try:
        df = df.with_columns(pl.col(raw_date_col).str.to_datetime(strict=False).dt.date().alias("det_date"))
        print(f"\nParsed date column '{raw_date_col}' -> det_date")
    except Exception as e:
        print(f"\nCould not parse date column '{raw_date_col}': {e}")
        raw_date_col = None

# Year/month stats
if raw_date_col:
    year_counts = df.group_by(pl.col("det_date").dt.year().alias("year")).len().sort("year")
    ym_counts = df.group_by(
        [pl.col("det_date").dt.year().alias("year"), pl.col("det_date").dt.month().alias("month")]
    ).len().sort(["year", "month"])
    print("\nCounts by year:")
    print(year_counts)
    print("\nCounts by year-month:")
    print(ym_counts.head(24))
else:
    print("\nNo date column found; skipping year/month stats.")

# Sex proportions
if sex_col:
    sex_total = df.group_by(sex_col).len().with_columns((pl.col("len") / pl.col("len").sum()).alias("proportion"))
    print("\nSex distribution (overall):")
    print(sex_total)
    if raw_date_col:
        sex_year = df.group_by([pl.col("det_date").dt.year().alias("year"), sex_col]).len()
        sex_year = sex_year.with_columns(
            (pl.col("len") / pl.sum("len").over("year")).alias("proportion")
        ).sort(["year", sex_col])
        print("\nSex distribution by year:")
        print(sex_year)
else:
    print("\nNo sex column found.")

# Age median using Birth Year
if birth_year_col:
    try:
        df = df.with_columns(pl.col(birth_year_col).cast(pl.Int64))
        if raw_date_col:
            # Use detention year for age calculation
            df = df.with_columns(pl.col("det_date").dt.year().alias("det_year"))
            df = df.with_columns((pl.col("det_year") - pl.col(birth_year_col)).alias("age"))
        else:
            # Fallback: age unknown without detention year
            df = df.with_columns(pl.lit(None).alias("age"))
        age_total = df.select(pl.col("age").median().alias("age_median"))
        print("\nAge median (overall):", age_total)
        if raw_date_col:
            age_year = df.group_by("det_year").agg(
                pl.col("age").median().alias("age_median")
            ).sort("det_year")
            print("\nAge median by year:")
            print(age_year)
    except Exception as e:
        print(f"\nCould not compute age median: {e}")
else:
    print("\nNo birth year column found (Birth Year).")

```
